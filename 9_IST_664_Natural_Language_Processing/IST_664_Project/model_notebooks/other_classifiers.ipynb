```python
featuresets_sen1 = [(sentiment_features(text), label) for text, label in zip(word_tokens, labels)]

train = []
train_label =[]
test = []
test_label=[]
for t,f in  zip(test_train,featuresets_sen1):
    if(t == "train"):
        train.append(f[0])
        train_label.append(f[1])
    else:
        test.append(f[0])
        test_label.append(f[1])
```
